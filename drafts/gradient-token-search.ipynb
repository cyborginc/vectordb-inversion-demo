```python
# Gradient-Based Token Search for Embedding Inversion
# This notebook optimizes probability distributions over vocabulary to invert embeddings

# %% [markdown]
# ## 1. Setup and Load Test Corpus

# %%
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModel
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
import torch.nn.functional as F
from tqdm import tqdm
import pandas as pd
import seaborn as sns

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Load model components
model_name = 'sentence-transformers/all-MiniLM-L6-v2'
sentence_model = SentenceTransformer(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
transformer_model = AutoModel.from_pretrained(model_name).to(device)

# Same diverse test corpus
test_corpus = [
    # News headlines
    "Scientists discover new species of deep-sea fish in Pacific Ocean",
    "Stock market reaches all-time high amid economic recovery",
    "Climate change accelerates Arctic ice melt, study finds",
    
    # Technical descriptions
    "The algorithm uses dynamic programming to optimize runtime complexity",
    "Machine learning model achieves 95% accuracy on test dataset",
    "Quantum computers leverage superposition for parallel processing",
    
    # Questions
    "What causes the northern lights to appear in the sky?",
    "How does photosynthesis convert sunlight into energy?",
    "Why do some materials conduct electricity better than others?",
    
    # Instructions
    "Mix flour, eggs, and milk to create pancake batter",
    "Press and hold the power button for 10 seconds to reset",
    "Apply two coats of paint, allowing each to dry completely",
    
    # Conversational
    "I really enjoyed the movie we watched last night",
    "The weather has been unusually warm for this time of year",
    "Let's meet at the coffee shop downtown at 3 PM",
    
    # Literary/Quotes
    "To be or not to be, that is the question",
    "The journey of a thousand miles begins with a single step",
    "All that glitters is not gold",
    
    # Facts
    "The human brain contains approximately 86 billion neurons",
    "Water boils at 100 degrees Celsius at sea level",
    "The speed of light in vacuum is 299,792,458 meters per second"
]

# Create embeddings
target_embeddings = sentence_model.encode(test_corpus)
print(f"Created {len(target_embeddings)} embeddings")

# %% [markdown]
# ## 2. Precompute Token Embeddings for Efficiency

# %%
# Precompute all token embeddings
print("Precomputing token embeddings...")
vocab_size = tokenizer.vocab_size

# Get all token embeddings (this is memory intensive but makes optimization faster)
all_token_ids = torch.arange(vocab_size).to(device)
with torch.no_grad():
    all_token_embeddings = transformer_model.embeddings.word_embeddings(all_token_ids)

print(f"Token embedding matrix shape: {all_token_embeddings.shape}")

# Create mapping of common tokens for analysis
common_tokens = ['the', 'of', 'and', 'to', 'in', 'is', 'for', 'that', 'with', 'on',
                'algorithm', 'computer', 'science', 'data', 'model', 'system',
                'climate', 'market', 'quantum', 'machine', 'learning']

token_to_id = {token: tokenizer.encode(token, add_special_tokens=False)[0] 
               for token in common_tokens if len(tokenizer.encode(token, add_special_tokens=False)) == 1}

# %% [markdown]
# ## 3. Implement Gradient-Based Token Search

# %%
class TokenSearchInverter:
    def __init__(self, model, tokenizer, token_embeddings, device):
        self.model = model
        self.tokenizer = tokenizer
        self.token_embeddings = token_embeddings
        self.device = device
        self.vocab_size = tokenizer.vocab_size
        
    def mean_pooling(self, model_output, attention_mask):
        token_embeddings = model_output[0]
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
        return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)
    
    def forward_with_soft_tokens(self, token_logits, temperature=1.0):
        """Forward pass using soft token probabilities"""
        # Apply temperature and softmax to get probabilities
        token_probs = F.softmax(token_logits / temperature, dim=-1)
        
        # Compute weighted embedding for each position
        # Shape: (batch, seq_len, vocab_size) x (vocab_size, embed_dim) -> (batch, seq_len, embed_dim)
        soft_embeddings = torch.matmul(token_probs, self.token_embeddings)
        
        # Add batch dimension if needed
        if soft_embeddings.dim() == 2:
            soft_embeddings = soft_embeddings.unsqueeze(0)
        
        # Create attention mask
        attention_mask = torch.ones(soft_embeddings.shape[0], soft_embeddings.shape[1]).to(self.device)
        
        # Forward through model
        outputs = self.model(inputs_embeds=soft_embeddings, attention_mask=attention_mask)
        sentence_embeddings = self.mean_pooling(outputs, attention_mask)
        
        # Normalize
        sentence_embeddings = F.normalize(sentence_embeddings, p=2, dim=1)
        
        return sentence_embeddings
    
    def invert_embedding(self, target_embedding, num_tokens=10, num_iterations=1000, 
                        lr=0.1, temperature_schedule='cosine', 
                        sparsity_weight=0.01, entropy_weight=0.001):
        """
        Invert embedding by optimizing token probability distributions
        """
        # Convert target to tensor
        target = torch.tensor(target_embedding, dtype=torch.float32).unsqueeze(0).to(self.device)
        target = F.normalize(target, p=2, dim=1)
        
        # Initialize token logits (including CLS and SEP)
        # Start with slight randomness
        token_logits = torch.randn(num_tokens + 2, self.vocab_size, device=self.device) * 0.01
        token_logits.requires_grad = True
        
        # Set CLS and SEP tokens with high confidence
        cls_id = self.tokenizer.cls_token_id
        sep_id = self.tokenizer.sep_token_id
        with torch.no_grad():
            token_logits[0, :] = -10.0  # Low scores for all
            token_logits[0, cls_id] = 10.0  # High score for CLS
            token_logits[-1, :] = -10.0
            token_logits[-1, sep_id] = 10.0  # High score for SEP
        
        # Optimizer
        optimizer = optim.Adam([token_logits], lr=lr)
        
        losses = []
        temperatures = []
        entropies = []
        
        for iteration in range(num_iterations):
            optimizer.zero_grad()
            
            # Temperature schedule
            if temperature_schedule == 'cosine':
                temperature = 0.5 + 0.5 * np.cos(np.pi * iteration / num_iterations)
            elif temperature_schedule == 'linear':
                temperature = 1.0 - 0.9 * (iteration / num_iterations)
            else:
                temperature = 1.0
            
            temperatures.append(temperature)
            
            # Forward pass
            predicted = self.forward_with_soft_tokens(token_logits, temperature)
            
            # Main loss: negative cosine similarity
            main_loss = 1 - F.cosine_similarity(predicted, target)
            
            # Sparsity loss: encourage one-hot distributions
            token_probs = F.softmax(token_logits / temperature, dim=-1)
            sparsity_loss = -torch.mean(torch.max(token_probs, dim=-1)[0])
            
            # Entropy loss: encourage low entropy (decisive choices)
            entropy = -torch.sum(token_probs * torch.log(token_probs + 1e-10), dim=-1)
            entropy_loss = torch.mean(entropy[1:-1])  # Exclude CLS/SEP
            entropies.append(entropy_loss.item())
            
            # Total loss
            loss = main_loss + sparsity_weight * sparsity_loss + entropy_weight * entropy_loss
            
            # Backward
            loss.backward()
            
            # Don't update CLS/SEP logits
            with torch.no_grad():
                token_logits.grad[0, :] = 0
                token_logits.grad[-1, :] = 0
            
            optimizer.step()
            
            losses.append(main_loss.item())
            
            if iteration % 200 == 0:
                print(f"Iteration {iteration}, Loss: {main_loss.item():.6f}, "
                      f"Temperature: {temperature:.3f}, Entropy: {entropy_loss.item():.3f}")
        
        # Extract final tokens
        with torch.no_grad():
            final_probs = F.softmax(token_logits / 0.1, dim=-1)  # Low temperature for final selection
            token_ids = torch.argmax(final_probs, dim=-1)
            
            # Also get top-k alternatives for each position
            top_k = 5
            top_probs, top_indices = torch.topk(final_probs[1:-1], k=top_k, dim=-1)
        
        # Decode tokens
        main_tokens = [self.tokenizer.decode([tid.item()]) for tid in token_ids[1:-1]]
        
        # Get alternatives
        alternatives = []
        for pos in range(len(main_tokens)):
            pos_alternatives = []
            for k in range(top_k):
                token = self.tokenizer.decode([top_indices[pos, k].item()])
                prob = top_probs[pos, k].item()
                pos_alternatives.append((token, prob))
            alternatives.append(pos_alternatives)
        
        return {
            'tokens': main_tokens,
            'alternatives': alternatives,
            'losses': losses,
            'temperatures': temperatures,
            'entropies': entropies,
            'final_probs': final_probs[1:-1].cpu().numpy()
        }

# %% [markdown]
# ## 4. Run Token Search on Test Corpus

# %%
# Initialize inverter
inverter = TokenSearchInverter(transformer_model, tokenizer, all_token_embeddings, device)

# Test on subset
results = []

for i, (text, embedding) in enumerate(zip(test_corpus[:10], target_embeddings[:10])):
    print(f"\n{'='*60}")
    print(f"Original: {text}")
    print(f"{'='*60}")
    
    # Run inversion
    result = inverter.invert_embedding(
        embedding,
        num_tokens=15,
        num_iterations=1000,
        lr=0.1,
        temperature_schedule='cosine',
        sparsity_weight=0.01
    )
    
    # Reconstruct text
    reconstructed = ' '.join(result['tokens'])
    
    # Verify similarity
    reconstructed_embedding = sentence_model.encode([reconstructed])[0]
    similarity = cosine_similarity([embedding], [reconstructed_embedding])[0][0]
    
    print(f"Reconstructed: {reconstructed}")
    print(f"Similarity: {similarity:.4f}")
    
    # Show top alternatives for first 5 positions
    print("\nToken alternatives (token: probability):")
    for pos, alts in enumerate(result['alternatives'][:5]):
        alt_str = ' | '.join([f"{token}:{prob:.3f}" for token, prob in alts[:3]])
        print(f"  Position {pos}: {alt_str}")
    
    results.append({
        'original': text,
        'reconstructed': reconstructed,
        'similarity': similarity,
        'result': result
    })

# %% [markdown]
# ## 5. Visualize Token Probability Evolution

# %%
# Visualize how token probabilities evolve during optimization
example_result = results[0]['result']

# Plot loss and temperature
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 10))

# Loss curve
ax1.plot(example_result['losses'])
ax1.set_title('Optimization Loss (1 - Cosine Similarity)')
ax1.set_xlabel('Iteration')
ax1.set_ylabel('Loss')
ax1.grid(True, alpha=0.3)

# Temperature schedule
ax2.plot(example_result['temperatures'], color='orange')
ax2.set_title('Temperature Schedule')
ax2.set_xlabel('Iteration')
ax2.set_ylabel('Temperature')
ax2.grid(True, alpha=0.3)

# Entropy
ax3.plot(example_result['entropies'], color='green')
ax3.set_title('Average Token Entropy')
ax3.set_xlabel('Iteration')
ax3.set_ylabel('Entropy')
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 6. Analyze Token Probability Distributions

# %%
# Visualize final probability distributions for one example
example_probs = results[0]['result']['final_probs']
example_tokens = results[0]['result']['tokens']

# Get top 10 tokens for each position
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for pos in range(min(6, len(example_tokens))):
    ax = axes[pos]
    
    # Get top 10 tokens by probability
    probs = example_probs[pos]
    top_indices = np.argsort(probs)[-10:][::-1]
    top_tokens = [tokenizer.decode([idx]) for idx in top_indices]
    top_probs = probs[top_indices]
    
    # Plot
    ax.barh(range(len(top_tokens)), top_probs)
    ax.set_yticks(range(len(top_tokens)))
    ax.set_yticklabels(top_tokens)
    ax.set_xlabel('Probability')
    ax.set_title(f'Position {pos}: "{example_tokens[pos]}"')
    ax.set_xlim(0, 1)

plt.tight_layout()
plt.show()

# %% [markdown]
# ## 7. Compare Different Temperature Schedules

# %%
# Test different temperature schedules
temp_schedules = ['cosine', 'linear', 'constant']
schedule_results = {}

test_text = test_corpus[4]  # Technical text
test_embedding = target_embeddings[4]

print(f"Testing temperature schedules on: {test_text}\n")

for schedule in temp_schedules:
    result = inverter.invert_embedding(
        test_embedding,
        num_tokens=10,
        num_iterations=500,
        temperature_schedule=schedule
    )
    
    reconstructed = ' '.join(result['tokens'])
    reconstructed_embedding = sentence_model.encode([reconstructed])[0]
    similarity = cosine_similarity([test_embedding], [reconstructed_embedding])[0][0]
    
    schedule_results[schedule] = {
        'text': reconstructed,
        'similarity': similarity,
        'final_entropy': result['entropies'][-1]
    }
    
    print(f"{schedule:10} | Similarity: {similarity:.4f} | Entropy: {result['entropies'][-1]:.4f}")
    print(f"            | {reconstructed}\n")

# %% [markdown]
# ## 8. Statistical Analysis Across Corpus

# %%
# Run on full corpus with consistent parameters
all_results = []

print("Running inversion on full corpus...")
for text, embedding in tqdm(zip(test_corpus, target_embeddings)):
    result = inverter.invert_embedding(
        embedding,
        num_tokens=12,
        num_iterations=500,
        lr=0.1,
        temperature_schedule='cosine'
    )
    
    reconstructed = ' '.join(result['tokens'])
    reconstructed_embedding = sentence_model.encode([reconstructed])[0]
    similarity = cosine_similarity([embedding], [reconstructed_embedding])[0][0]
    
    all_results.append({
        'original': text,
        'reconstructed': reconstructed,
        'similarity': similarity,
        'text_length': len(text.split()),
        'category': text.split()[0]  # Rough categorization
    })

# Create DataFrame
df = pd.DataFrame(all_results)

# Statistical summary
print("\nOverall Statistics:")
print(f"Mean similarity: {df['similarity'].mean():.4f} ± {df['similarity'].std():.4f}")
print(f"Min similarity: {df['similarity'].min():.4f}")
print(f"Max similarity: {df['similarity'].max():.4f}")

# Plot similarity distribution
plt.figure(figsize=(10, 4))
plt.hist(df['similarity'], bins=20, alpha=0.7, edgecolor='black')
plt.xlabel('Cosine Similarity')
plt.ylabel('Count')
plt.title('Distribution of Reconstruction Similarities')
plt.grid(True, alpha=0.3)
plt.show()

# %% [markdown]
# ## 9. Failure Analysis

# %%
# Analyze worst reconstructions
worst_results = df.nsmallest(5, 'similarity')

print("Worst reconstructions:")
for idx, row in worst_results.iterrows():
    print(f"\nSimilarity: {row['similarity']:.4f}")
    print(f"Original:      {row['original']}")
    print(f"Reconstructed: {row['reconstructed']}")

# Analyze best reconstructions
best_results = df.nlargest(5, 'similarity')

print("\n\nBest reconstructions:")
for idx, row in best_results.iterrows():
    print(f"\nSimilarity: {row['similarity']:.4f}")
    print(f"Original:      {row['original']}")
    print(f"Reconstructed: {row['reconstructed']}")

# %% [markdown]
# ## 10. Conclusions and Security Implications

# %%
print("=== GRADIENT-BASED TOKEN SEARCH RESULTS ===\n")

print("ADVANTAGES OVER CONTINUOUS OPTIMIZATION:")
print("1. Always produces valid token sequences")
print("2. Can sample multiple reconstructions from final distribution")
print("3. Provides confidence scores for each token")
print("4. More interpretable optimization process")

print("\nKEY FINDINGS:")
print(f"- Average reconstruction similarity: {df['similarity'].mean():.4f}")
print("- Temperature scheduling crucial for convergence")
print("- Technical/unique terms easier to recover")
print("- Short texts generally harder to reconstruct accurately")

print("\nSECURITY IMPLICATIONS:")
print("- Gradient-based search can recover semantic content")
print("- Unique terminology and patterns are vulnerable")
print("- Even 'failed' inversions reveal information about text type")
print("- Probability distributions show alternative interpretations")

print("\n⚠️  CRITICAL: Both optimization methods demonstrate that embeddings")
print("leak substantial information and should not be considered secure!")
```
